```python
from tokenizers import Tokenizer, trainers, models, pre_tokenizers, processors, normalizers
from tokenizers.normalizers import Lowercase, NFKC
from tokenizers.pre_tokenizers import Whitespace, Punctuation
from tokenizers.processors import TemplateProcessing
```

```python
# 1. Initialize empty BPE tokenizer

tokenizer = Tokenizer(models.BPE())

# 2. Normalization rules
tokenizer.normalizer = normalizers.Sequence([
    NFKC(),       # normalize unicode
    # Lowercase()  # OPTIONAL — uncomment if you want lowercase tokenizer
])

# 3. Pre-tokenization rules

tokenizer.pre_tokenizer = pre_tokenizers.Sequence([
    Whitespace(),
    Punctuation()
])

# 4. BPE Trainer configuration

trainer = trainers.BpeTrainer(
    vocab_size=32000,
    min_frequency=2,
    special_tokens=[
        "<pad>", "<unk>", "<s>", "</s>"
    ]
)

# 5. Train tokenizer on your LM corpus

tokenizer.train(
    files=["prepared/lm_corpus.txt"],
    trainer=trainer
)

# 6. Add post-processing template

tokenizer.post_processor = TemplateProcessing(
    single="<s> $A </s>",
    pair="<s> $A </s> </s> $B </s>",
    special_tokens=[
        ("<s>", tokenizer.token_to_id("<s>")),
        ("</s>", tokenizer.token_to_id("</s>"))
    ]
)

# 7. Save tokenizer

tokenizer.save("tokens/tokenizer.json")

print("Tokenizer training completed")
```
